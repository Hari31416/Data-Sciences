```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def change(s):
    if s=='male':
        s = 1
    else:
        s = 0
    return s

test = pd.read_csv('test.csv')
test = test.drop(['Name', 'Ticket', 'Cabin'], axis = 1)
test['Sex'] = test['Sex'].apply(change)
```

```python
df3 = test.groupby(['Pclass', 'Sex']).mean()
```

```python
df4 = test[test['Age'].notnull()]
```

```python
classes = [1,2,3]
Sex = [1, 0]
for c in classes:
    for s in Sex:
        df_new = test[(test['Pclass'] == c) & (test['Sex'] == s)].fillna(df3.loc[(c,s)]['Age'])
        df4 = pd.concat([df4, df_new], axis = 0)
```

```python
df4
```

```python
df5 = df4.drop_duplicates()
final_df = df5.sort_index()
```

```python
final_df
```

```python
test_str = pd.get_dummies(final_df['Embarked'], drop_first=True)
```

```python
test = final_df.drop('Embarked', axis = 1)
```

```python
test = pd.concat([test, test_str], axis = 1)
```

```python
test.to_csv('test_final.csv', index = False)
```
