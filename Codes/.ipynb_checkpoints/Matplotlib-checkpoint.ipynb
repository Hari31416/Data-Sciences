```python
import numpy as np
x = np.linspace(-2, 5, 100)
y = x ** 2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0, 1, 1])
axes1.set_xlabel('first axis')
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])
axes2.set_xlabel('second axis')
axes1.plot(x,y, 'r')
axes2.plot(y,x, 'g')
fig.savefig('fig.png', dpi = 300)
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2, figsize=(16,5))
axes[0].set_xlabel('x')
axes[0].set_ylabel('$x^2$ and $e^x$')

axes[1].set_xlabel('x')
axes[1].set_ylabel('$\sin{x}$')

axes[0].plot(x,x**3, 'r', lw = 2, label='$x^3$')
axes[0].plot(x,np.exp(x), 'g--', label = "$e^x$")

axes[1].plot(x,np.sin(x), 'r', label = '$\sin{x}$')
axes[0].legend(fontsize= 15, loc = 0)
```

```python
fig = plt.figure(figsize= (15,15))
ax1 = plt.subplot2grid((3,3), (0,0))
ax2 = plt.subplot2grid((3,3), (0,1), rowspan = 2)
ax3 = plt.subplot2grid((3,3), (0,2), rowspan = 3)
ax4 = plt.subplot2grid((3,3), (1,0), rowspan = 2)
ax5 = plt.subplot2grid((3,3), (2,1))

ax1.plot(x, x**2, color = 'red', label = '$x^2$')
ax2.plot(x, x**6, color = 'blue', label = '$x^6$')
ax3.plot(x, x**4, color = 'green', label = '$x^4$')
ax4.plot(x, x**5, color = 'black', label = '$x^5$')
ax5.plot(x, x**3, color = 'indigo', label = '$x^3$')

ax1.set_title('$x^2$')
ax2.set_title('$x^6$')
ax3.set_title('$x^4$')
ax4.set_title('$x^5$')
ax5.set_title('$x^3$')
fig.savefig('fig.png', dpi = 200)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some mock data
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:green'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 1000)
f = np.arange(0, 1000)
g = np.sin(np.arange(0, 10, 0.01) * 2) * 1000

plt.plot(x, f, '-')
plt.plot(x, g, '-')

idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
plt.plot(x[idx], f[idx], 'ro')
print(x[idx], f[idx])
plt.show()
```
